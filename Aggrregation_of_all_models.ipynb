```python
import pandas as pd
import numpy as np
import time as t
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
dataframe_project = pd.read_csv(r'E:\5th sem\ML\lab\patches_gabor_15816_1.csv')
dataframe_project
```

```python
input_project = dataframe_project.loc[:, dataframe_project.columns[2:]]
output_project = dataframe_project.loc[:, dataframe_project.columns[1]]

input_train, input_test, output_train, output_test = train_test_split(input_project, output_project, test_size=0.3, random_state=42)
```

```python
# Project data classification rules
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
randomForest_model = RandomForestClassifier()
scores_rf = cross_val_score(randomForest_model, input_project.values, output_project.values, cv=10)
start_train_rf = t.time()
grid_search_rf = GridSearchCV(randomForest_model, param_grid={'n_estimators': [10, 20, 30, 50, 100]}, cv=10)
grid_search_rf.fit(input_train, output_train)
end_train_rf = t.time()
training_time_rf = end_train_rf - start_train_rf
predicted_rf_train = grid_search_rf.predict(input_train.values)
start_test_rf = t.time()
predicted_rf_test = grid_search_rf.predict(input_test.values)
end_test_rf = t.time()
testing_time_rf = end_test_rf - start_test_rf
```

```python
print("Training data")
print(classification_report(output_train, predicted_rf_train))
print("Accuracy: ", accuracy_score(output_train, predicted_rf_train))
print("Precision: ", precision_score(output_train, predicted_rf_train, average='weighted'))
print("Recall: ", recall_score(output_train, predicted_rf_train, average='weighted'))
print("F1-score: ", f1_score(output_train, predicted_rf_train, average='weighted'))
print(training_time_rf, 's')
print("Testing data")
print(classification_report(output_test, predicted_rf_test))
print("Accuracy: ", accuracy_score(output_test, predicted_rf_test))
print("Precision: ", precision_score(output_test, predicted_rf_test, average='weighted'))
print("Recall: ", recall_score(output_test, predicted_rf_test, average='weighted'))
print("F1-score: ", f1_score(output_test, predicted_rf_test, average='weighted'))
print(testing_time_rf, 's')
print(grid_search_rf.best_params_)
print('Mean accuracy: ',scores_rf.mean())
print('Standard deviation: ',scores_rf.std())
```
