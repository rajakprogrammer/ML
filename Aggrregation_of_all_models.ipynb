```python
import pandas as pd
import numpy as np
import time as t
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
dataframe_project = pd.read_csv(r'D:\Kartik Projects\Semester 5\Machine Learning\2023-09-12\Custom_CNN_Features.csv')
dataframe_project
```

```python
input_project = dataframe_project.loc[:, dataframe_project.columns[2:]]
output_project = dataframe_project.loc[:, dataframe_project.columns[1]]

input_train, input_test, output_train, output_test = train_test_split(input_project, output_project, test_size=0.3, random_state=42)
```

```python
# Project data classification rules
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
start_train_knn = t.time()
knn_model.fit(input_train, output_train)
end_train_knn = t.time()
training_time_knn = end_train_knn - start_train_knn
predicted_knn_train = knn_model.predict(input_train.values)
start_test_knn = t.time()
predicted_knn_test = knn_model.predict(input_test.values)
end_test_knn = t.time()
testing_time_knn = end_test_knn - start_test_knn
```

```python
print("Training data")
print(classification_report(output_train, predicted_knn_train))
print("Accuracy: ", accuracy_score(output_train, predicted_knn_train))
print("Precision: ", precision_score(output_train, predicted_knn_train, average='weighted'))
print("Recall: ", recall_score(output_train, predicted_knn_train, average='weighted'))
print("F1-score: ", f1_score(output_train, predicted_knn_train, average='weighted'))
print(training_time_knn, 's')
print("Testing data")
print(classification_report(output_test, predicted_knn_test))
print("Accuracy: ", accuracy_score(output_test, predicted_knn_test))
print("Precision: ", precision_score(output_test, predicted_knn_test, average='weighted'))
print("Recall: ", recall_score(output_test, predicted_knn_test, average='weighted'))
print("F1-score: ", f1_score(output_test, predicted_knn_test, average='weighted'))
print(testing_time_knn, 's')
```
