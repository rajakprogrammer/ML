```python
# Question A1:
# Calculate the prior probabilities of each class values

# Creating the customer data table
table = {
    'age': ['<=30', '<=30', '31...40', '>40', '>40', '>40', '31...40', '<=30', '<=30', '>40', '<=30', '31...40', '31...40', '>40'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

dataframe_customer = pd.DataFrame(table)
dataframe_customer
```

```python
# Calculating the probability of each class

# Calculating the prior probabitity of the class yes
probability_yes = dataframe_customer.loc[dataframe_customer['buys_computer'] == 'yes', 'buys_computer'].count() / dataframe_customer.loc[:, 'buys_computer'].count()

# Calculating the prior probability of the class no
probability_no = dataframe_customer.loc[dataframe_customer['buys_computer'] == 'no', 'buys_computer'].count() / dataframe_customer.loc[:, 'buys_computer'].count()

print("The prior probabilities of class yes is: ", probability_yes)
print("The prior probabilities of class no is: ", probability_no)
```

```python
# Question A2:
# Calculate the class densities of all the features given the various classes

# Creating the dataframes for age and class label
dataframe_age_yes = pd.DataFrame(dataframe_customer.loc[dataframe_customer['buys_computer'] == 'yes', dataframe_customer.columns[[0, 4]]])

dataframe_age_no = pd.DataFrame(dataframe_customer.loc[dataframe_customer['buys_computer'] == 'no', dataframe_customer.columns[[0, 4]]])

# Probabities for each age value given the class is yes
# P(<=30|yes)
probability_less_equal_30_given_yes = dataframe_age_yes.loc[dataframe_age_yes['age'] == '<=30', 'age'].count() / dataframe_age_yes.loc[:, 'buys_computer'].count()
print("P(<=30 | yes) = ", probability_less_equal_30_given_yes)
# P(31...40|yes)
probability_31_to_40_given_yes = dataframe_age_yes.loc[dataframe_age_yes['age'] == '31...40', 'age'].count() / dataframe_age_yes.loc[:, 'buys_computer'].count()
print("P(31...40 | yes) = ", probability_31_to_40_given_yes)
# P(>40|yes)
probability_great_than_40_given_yes = dataframe_age_yes.loc[dataframe_age_yes['age'] == '>40', 'age'].count() / dataframe_age_yes.loc[:, 'buys_computer'].count()
print("P(>40 | yes) = ", probability_great_than_40_given_yes)

# Probabilities for each age value given the class is no
# P(<=30|no)
probability_less_equal_30_given_no = dataframe_age_no.loc[dataframe_age_no['age'] == '<=30', 'age'].count() / dataframe_age_no.loc[:, 'buys_computer'].count()
print("P(<=30 | no) = ", probability_less_equal_30_given_no)
# P(31...40|no)
probability_31_to_40_no = dataframe_age_no.loc[dataframe_age_no['age'] == '31...40', 'age'].count() / dataframe_age_no.loc[:, 'buys_computer'].count()
print("P(31...40 | no) = ", probability_31_to_40_no)
# P(>40|no)
probability_great_than_40_given_no = dataframe_age_no.loc[dataframe_age_no['age'] == '>40', 'age'].count() / dataframe_age_no.loc[:, 'buys_computer'].count()
print("P(>40 | no) = ", probability_great_than_40_given_no)
```

```python
# Applying the label encoding each of the features
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
for i in dataframe_customer.columns:
    dataframe_customer[i] = label_encoder.fit_transform(dataframe_customer[i])
dataframe_customer
```

```python
# Question A4:
# Apply the Naive Bayes classifier for the above table data
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

dataframe_customer_input = dataframe_customer.iloc[:, :4]
dataframe_customer_output = dataframe_customer.iloc[:, 4]

# Splitting the data into training data and testing data
customer_input_train, customer_input_test, customer_output_train, customer_output_test = train_test_split(dataframe_customer_input, dataframe_customer_output, test_size=0.3, random_state=42)

# Applying the naive bayes classifier on customer data
model_naive_bayes = GaussianNB()
model_naive_bayes.fit(customer_input_train, customer_output_train)

customer_output_predicted = model_naive_bayes.predict(customer_input_test)
accuracy_customer_data = accuracy_score(customer_output_test, customer_output_predicted)
print("The accuracy of the customer data is: ", accuracy_customer_data)
```

```python
# Question A5:
# Apply the Naive Bayes classifier for the project data
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
dataframe_project_input = dataframe_project.iloc[:, 2:]
dataframe_project_output = dataframe_project.iloc[:, 1]

# Splitting the project data into training data and testing data
project_input_train, project_input_test, project_output_train, project_output_test = train_test_split(dataframe_project_input, dataframe_project_output, test_size=0.3, random_state=42)

# Applying the naive bayes classifier on project data
model_project_naive_bayes = GaussianNB()
model_project_naive_bayes.fit(project_input_train, project_output_train)

project_output_predicted = model_project_naive_bayes.predict(project_input_test)
accuracy_project_data = accuracy_score(project_output_test, project_output_predicted)
print("The accuracy of the project data is: ", accuracy_project_data)
print("The precision score of the project data is:", precision_score(project_output_test, project_output_predicted, average='weighted'))
print("The recall score of the project data is:", recall_score(project_output_test, project_output_predicted, average='weighted'))
print("The f1-score of the project data is:", f1_score(project_output_test, project_output_predicted, average='weighted'))
print(classification_report(project_output_test, project_output_predicted))
```
