```python
dataframe_project = pd.read_csv(r'E:\5th sem\ML\lab\Custom_CNN_Features.csv')
dataframe_project
```

```python
# Question A1:Merge your train & test sets and remove the class labels. Use k-means algorithm with k = 3 or 5 
#  (based on your dataset) to form the clusters.from sklearn.cluster import KMeans

# Selecting columns starting from the 5th column (index 4) as input data
input_data =dataframe_project.iloc[:, 4:]
# Specify the number of clusters (in this case, 3)
cluster_model = KMeans(n_clusters=3)
cluster_model.fit(input_data)
cluster_model.labels_
```

```python
# Question A2: Create a linear regression model on this data. Consider using one feature as independent variable while the other as dependent variable (you may also round this number to integer). After 
#   the model is created, calculate the mean square error by predicting the values from the model. 

from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

# Assuming 'input_data' is your input data for KMeans clustering

# Range of k values to test
k_values = range(1, 32)

# List to store the sum of squared distances for each k
euclidean_distances = []

# Iterate through different k values
for k in k_values:
    # Create KMeans model with the current k value
    k_means_model = KMeans(n_clusters=k)
    
    # Fit the model to the input data
    k_means_model.fit(input_data)
    
    # Append the sum of squared distances (inertia) to the list
    euclidean_distances.append(k_means_model.inertia_)

# Plotting the Elbow curve
plt.plot(k_values, euclidean_distances, 'bx-')
plt.xlabel('K values')
plt.ylabel('Euclidean distances')
plt.title('Elbow curve')
plt.show()
```

```python
# Question A3: Study all the parameters and attributes associated with linear regression model.
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as hierarchy
import matplotlib.pyplot as plt

# Assuming 'input_data' is your input data for Agglomerative Clustering

# Create an AgglomerativeClustering model
agglomerative_model = AgglomerativeClustering()

# Fit the model to the input data
agglomerative_model.fit(input_data)

# Generate the linkage matrix using Ward method
linkage_matrix = hierarchy.linkage(input_data, method='ward')

# Create a dendrogram using the linkage matrix
dendrogram = hierarchy.dendrogram(linkage_matrix)

# Set plot labels
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Euclidean Distance')

# Display the dendrogram
plt.show()
```

```python
# Question A4: 4. Using the training set available for your project, train a logistic regression classifier. Use this classifier to evaluate your test set accuracy. Study the various parameters associated with logistic 
#   regression model and the role they play in the model training.

from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
# Selecting input features and output variable
dataframe_input = dataframe_project.iloc[:, 2:]
dataframe_output = dataframe_project.iloc[:, 1]
# Splitting the data into training and testing sets
input_train, input_test, output_train, output_test = train_test_split(dataframe_input, dataframe_output, test_size=0.3, random_state=42)
# Creating a Decision Tree Classifier
dec_tree_trained_model = DecisionTreeClassifier()
# Initializing SequentialFeatureSelector
feature_selection = SequentialFeatureSelector(dec_tree_trained_model, cv=5, n_features_to_select=7, scoring='accuracy')
feature_selection.fit(input_train, output_train)
# Getting the indices of the selected features
selected_features_indices = feature_selection.get_support(indices=True)
# Displaying the selected feature indices
print("The most prominent feature indices: ", selected_features_indices)
```

```python
# Question A5: Use a Regression Tree and k-NN Regressor for value prediction. Use the data employed for A1.
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

# Create a PCA model
pca_model = PCA()
# Fit the PCA model to the input data
pca_model.fit(input_data)
# Calculate the cumulative sum of explained variance ratios
variance_cum_sum = np.cumsum(pca_model.explained_variance_ratio_)
# Determine the number of principal components to retain for 95% variance
K_columns = np.argmax(variance_cum_sum >= 0.95) + 1
print("The number of principal components is: ", K_columns)
pca_k = PCA(n_components=K_columns)
# Transform the input data using the selected number of principal components
input_pca = pd.DataFrame(pca_k.fit_transform(input_data))
# Split the data into training and testing sets
input_pca_train, input_pca_test, output_pca_train, output_pca_test = train_test_split(input_pca, dataframe_output, test_size=0.3, random_state=42)
# Create a Decision Tree Classifier
dec_tree_trained_model = DecisionTreeClassifier()
dec_tree_trained_model.fit(input_pca_train, output_pca_train)
output_pca_pred = dec_tree_trained_model.predict(input_pca_test)
# Calculate and print the accuracy of the model with PCA
print('Accuracy of ', K_columns, ' components is: ', accuracy_score(output_pca_test, output_pca_pred))
```
