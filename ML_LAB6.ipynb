```python
# Question: A1. Take any 2 features from your project having has real numeric values.
features1 = df['f74'] # Taking column f74 as the first real values features
features2 = df['f59'] # Taking column f59 as the second real values features
print(features1)
print()
print(features2)
```

```python
# Question : A2. Creating a linear regression model on the given data
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import pandas as pd

# Assuming you have two DataFrames, features1 and features2
features_independent = pd.DataFrame(features1)  # Independent features
features_dependent = pd.DataFrame(features2)    # Dependent features

# Splitting the data into training and testing sets
features_independent_train, features_independent_test, features_dependent_train, features_dependent_test = train_test_split(
    features_independent, features_dependent, test_size=0.3, random_state=42)

# Create a Linear Regression model
regressor = linear_model.LinearRegression()

# Fit the model to the training data
regressor.fit(features_independent_train, features_dependent_train)

# Predicting the values using the test data
features_dependent_pred_lin_reg = regressor.predict(features_independent_test)
```

```python
# calculate the mean square error by predicting the values from the model.
from sklearn.metrics import mean_squared_error
    
# Assuming you have predicted values in features_dependent_pred_lin_reg and actual values in features_dependent_test
mean_square_error_features_lin_reg = mean_squared_error(features_dependent_test, features_dependent_pred_lin_reg)

print("The mean square error of the given dataset is:", mean_square_error_features_lin_reg)

```

```python
# Question: A3. Study all parameters and attributes associated with linear regression model
theta = np.sum((features_independent - features_independent.mean()) * (features_independent - features_independent.mean())) / np.sum((features_independent - features_independent.mean()) ** 2)
print("The intercept of the regression line is: ", regressor.intercept_)
print("The slope of the regression line is: ", theta)
```
